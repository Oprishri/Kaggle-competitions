# Extending features of train data

```python
def features1(data):
    drop_lst = ['casual', 'registered']
    data = data.drop(drop_lst,axis=1)
    data['datetime'] = pd.to_datetime(data['datetime'])
    data['dow'] =  data['datetime'].dt.dayofweek
    data['month'] =  data['datetime'].dt.month
    data['week'] = data['datetime'].dt.week
    data['hour'] = data['datetime'].dt.hour
    data['year'] =  data['datetime'].dt.year
    data['day'] =  data['datetime'].dt.day
    data = data.set_index(data['datetime'])
    data =  data.drop(labels='datetime', axis=1)
    data = pd.get_dummies( data, columns=['weather'])
    data =  data.drop(labels='weather_4', axis=1)
    data['temp_weath_1'] =  data['temp'] *  data['weather_1']
    data['temp_weath_2'] =  data['temp'] *  data['weather_2']
    data['temp_weath_3'] =  data['temp'] *  data['weather_3']
    data['temp_weath_1'] =  data['temp_weath_1'].astype(int)
    data['temp_weath_2'] =  data['temp_weath_2'].astype(int)
    data['temp_weath_3'] =  data['temp_weath_3'].astype(int)
    return(data)
```

```python
train=features1(train)
train.head()
```

# Extending features of test data

```python
def features2(data):
    data['datetime'] = pd.to_datetime(data['datetime'])
    data['dow'] =  data['datetime'].dt.dayofweek
    data['month'] =  data['datetime'].dt.month
    data['week'] = data['datetime'].dt.week
    data['hour'] = data['datetime'].dt.hour
    data['year'] =  data['datetime'].dt.year
    data['day'] =  data['datetime'].dt.day
    data = data.set_index(data['datetime'])
    data =  data.drop(labels='datetime', axis=1)
    data = pd.get_dummies( data, columns=['weather'])
    data =  data.drop(labels='weather_4', axis=1)
    data['temp_weath_1'] =  data['temp'] *  data['weather_1']
    data['temp_weath_2'] =  data['temp'] *  data['weather_2']
    data['temp_weath_3'] =  data['temp'] *  data['weather_3']
    data['temp_weath_1'] =  data['temp_weath_1'].astype(int)
    data['temp_weath_2'] =  data['temp_weath_2'].astype(int)
    data['temp_weath_3'] =  data['temp_weath_3'].astype(int)
    return(data)
```

```python
test=features2(test)
test.head()
```

```python
X = train.iloc[:, train.columns != 'count']
y = train['count']
```

# Random forest

```python
from sklearn.ensemble import RandomForestRegressor
def r_f():
    rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
    rf.fit(X,y)
    predictions = rf.predict(test)
    test['count']=predictions
    accuracy = rf.score(test.iloc[:, test.columns != 'count'],test['count'])
    return(accuracy, predictions)
```

```python
pred = r_f()
pred
```

# CSV

```python
def CSV():
    df=pd.DataFrame(pred[1], t['datetime'])
    df.columns=['count']
    df.to_csv(r'C:\solution.csv')
    return df
```
